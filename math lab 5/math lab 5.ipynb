```python
"Упражнение 2"
import numpy as np
compNum = [(4-5j)*(5-6j**3), (1+1j)**15]
def compCalc(compNum):
    print(
        f'Вычисления для комплексного числа{compNum}'
          ,np.real(compNum)
          ,np.imag(compNum)
          ,np.abs(compNum)
          ,np.conj(compNum)
          ,sep='\n'
          )

for item in compNum:
    compCalc(item)
```

```python
"Упражнение 3"
import matplotlib.pyplot as plt
import numpy as np
z1 = 1+3j 
z2 = 3+4j
z3 = z1+z2
z4 = z1-z2
z = np.array([z1,z2,z3,z4])

orig = np.zeros(4)
plt.plot()
q = plt.quiver(orig,orig, np.real(z), np.imag(z), angles='xy', scale_units='xy', scale=1, color="black")
plt.axis([-4,6,-4,9])
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title('Complex numbers')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
num = (2*np.sqrt(3)-2j)**1/8
r = np.abs(num)
t = np.linspace(0,2*np.pi,100)
z = r*(np.cos(t)+1j*np.sin(t))
plt.plot(np.real(z),np.imag(z),'r:')
plt.grid(True)
plt.axis('equal')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.show()
```

```python
z = 2*(3**0.5) - 2j
r = np.abs(z)
phi=np.angle(z)
k = np.arange(1,10)
zroot = r**(1/8)*( np.cos((phi+2.0*np.pi*k)/8) +1j*np.sin((phi+2.0*np.pi*k)/8))
plt.title('$\sqrt[8]{2*sqrt(3)-2j}$')
plt.plot(np.real(zroot), np.imag(zroot), ':r')
plt.axhline(y=0, color='black')
plt.axvline(x=0, color='black')
plt.axis('equal')
t = np.linspace(0,2*np.pi,100)
plt.grid()
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.plot(r**(1/8)*np.cos(t),r**(1/8)*np.sin(t),'--k')
```
