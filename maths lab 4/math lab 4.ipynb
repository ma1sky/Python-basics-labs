```python
"Task 1"

import matplotlib.pyplot as plt
import numpy as np
A = np.array([-2,0])
B = np.array([1,2])
C = np.array([1,-1])
AB = B - A
BC = C - B
AC = C - A
BA = A - B
BD = BA + BC
D = BD + B
AD = D - A
CD = D - C

plt.grid()
plt.ylim(-4, 3)
plt.xlim(-3, 2)
plt.arrow(-2, 0, AB[0], AB[1], width=0.03, length_includes_head=True, facecolor='Yellow')
plt.arrow(1, 2, BC[0], BC[1], width=0.03, length_includes_head=True, facecolor='Red')
plt.arrow(-2, 0, AC[0], AC[1], width=0.03, length_includes_head=True, facecolor='Blue')
plt.arrow(1, 2, BD[0], BD[1], width=0.03, length_includes_head=True, facecolor='Green')
plt.arrow(C[0], C[1], CD[0], CD[1], width=0.03, length_includes_head=True, facecolor='Pink')
plt.arrow(A[0], A[1], AD[0], AD[1], width=0.03, length_includes_head=True, facecolor='Violet')
plt.show()
```

```python
"Task 2"
import matplotlib.pyplot as plt
import numpy as np

a = np.array([-2,0])
b = np.array([3,4])
c = np.array([4,-1])
AB = b - a
BC = c - b
BA = a - b
AC = c - a
CA = a - c
CB = b - c

def calcCos(v1, v2):
    return np.dot(v1, v2) / np.linalg.norm(v1) / np.linalg.norm(v2)

def calcAngle(v1,v2):
    return round(np.degrees(np.arccos(calcCos(v1,v2))))
    
print('Проверка суммы величин углов в треугольнике: ',calcAngle(AB, AC) + calcAngle(BA, BC) + calcAngle(CA,CB))
print('Угол между AB и AC ',calcAngle(AB, AC))
print('Угол между BA и BC ',calcAngle(BA, BC))
print('Угол между CA и CB ', calcAngle(CA, CB))

plt.grid()

plt.arrow(a[0], a[1], AB[0], AB[1], width=0.03, length_includes_head=True, facecolor='Yellow')
plt.arrow(b[0], b[1], BC[0], BC[1], width=0.03, length_includes_head=True, facecolor='Yellow')
plt.arrow(c[0], c[1], CA[0], CA[1], width=0.03, length_includes_head=True, facecolor='Yellow')


plt.show()
```

```python
"Упражнение 5"
import numpy as np

def calcSin(v1, v2):
    return np.sqrt(1 - calcCos(v1, v2)**2)


def calcCos(v1, v2):
    return np.dot(v1, v2) / np.linalg.norm(v1) / np.linalg.norm(v2)

def calcArea(v1, v2):
    return 1/2 * (np.linalg.norm(v1)*np.linalg.norm(v2)*calcSin(v1, v2))

A = np.array([1,3,-1])
B = np.array([2,-1,4])
C = np.array([5,0,3])

AB = B - A
AC = C - A

print('Площадь треугольника = ', round(calcArea(AB, AC)))
```

```python
"Упражнение 6"
import numpy as np

def calcVol(matrix):
    return 1/6 * np.linalg.det(matrix)

AB = np.array([3, 4, 0])
AC = np.array([-3, 0, -1])
AD = np.array([0, 2, 5])
matrix = [AB, AC, AD]
print('Объем пирамиды = ',round(calcVol(matrix)))
```
